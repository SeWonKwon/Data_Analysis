## 함수적 종속성(Functional Dependency)

* 속성 X, Y에 대하여,<br>$x_{1} = x_2 (x_1, x_2 \in X)$이면, $ y_1 = y_2 (y_1, y_2 \in Y)$ <br>임을 보장할 때,속성 Y는 함수적으로 종속됨   


* X→Y로 표기
* X는 결정자(determinant)
* Y는 의존자(dependent)
* X가 Y를 함수적으로 결정
* Y가 X에 함수적으로 종속

## 유도 속성(derived attribute)
<br>
* 어떤 속성은 다른 속성이나 속성의 집합으로부터 유도 가능 <br>
<br>
* 연령 속성은 생년월일 속성의 유도 속성    <br>
<br>
* 총점 속성은 각 과목 점수 속성 집합의 유도 속성    <br>
<br>
* 일반적으로 데이터베이스 조회 성능을 높이기 위해 반정규화 과정에서 유도속성 정의   <br>
<br>
* 반정규화(de-normalization)    <br>
<br>
* 정규화(normalization)의 반대 개념으로 유도 속성의 정의    <br>
<br>
* 릴레이션의 수직/수평 분할 등이 존재    <br>
<br>
* 일반적으로 정규화는 데이터 중복을 제거하는 과정   <br>
<br>
* 반정규화는 성능 향상을 위해 중복을 허용하는 과정    <br>
<br>
* 유도 속성은 성능 향상에 도움이 되지만, 데이터 중복을 허용하는 과정으로 데이터 불일치 문제 발생<br>
<br>
* 월소득 속성값이 100만원인데 연소득 속성값이 1,000만원 (둘중 하나의 속성값에 오류가 존재)<br>

## 정규화되지 않은 테이블

* 정규화되지 않은 테이블의 사용은 데이터 중복의 원인이 됨<br>
<br>
* 일부러 정규화하지 않은 채 테이블을 사용하는 경우, <br>조인을 피함으로 성능을 향상시키지만, <br>이러한 테이블 구조는 데이터 중복으로 인한 데이터 일관성 저해 문제 발생<br>
<br>
* (예) 이름과 주소 정보를 키로 사용하여 구매자 주문을 저장하는 데이터베이스<br>
<br>
    * 같은 이름에 다른 주소를 갖는 경우에 대해 데이터 불일치 발생<br>
    <br>
    * 정규화 관점에서 구매자 주소는 구매자의 구매 품목과는 별개의 테이블에 저장 필요<br>
    <br>
    * 구매 테이블의 키가 (구매자 번호 + 구매 일시)라고 가정할 때,<br> 주소는 키에 부분 함수적 종속 관계에 있으므로 제2정규화 원칙에 따라 <br>구매자 번호와 함께 별도의 테이블로 분리 설계해야 함<br>
    <br>
    * 그러나 조회 성능 향상을 위해 테이블을 분리하지 않았을 때에는 <br>같은 구매자 번호에 대해서 다른 주소가 존재할 가능성은 언제든지 있음

## 상관계수(Correlation Coefficient)

* <p>  수치 속성에 대해서 속성 A와 B의 상관계수
<div>  </div></p>

 
* N: 튜플의 개수

* $a_i, b_i$: 튜플 i에서의 속성 A, B의 값

* $\bar{A}, \bar{B} :$ 속성 A,B의 평균값

* $\sigma_{A}, \sigma_{B}$ : 속성 A, B의 표준편차

* $-1<= \gamma_{A,B} <= +1$: 상관계수 결과 값 범위

* $\gamma_{A,B} \geqslant 0$ : 속성 A, B는 양의 상관관계positively correlated, 즉 B 값이 증가함에 따라 A 값이 증가
* $\gamma_{A,B} \leqslant 0$ : 속성 A, B는 음의 상관관계negatively correlated, 즉 B 값이 증가함에 따라 A 값이 감소
* $\gamma_{A,B} = 0$ : 속성 A, B는 독립적이며 둘 사이에 상관관계가 없음

>* 상관계수 $\gamma_{A,B}$의 절대값이 클수록 더 강한 상관성이 있다는 의미
>* 일반적으로 두 속성 간 상관성이 클수록 중복 속성으로 판단해서 하나의 속성을<br> 분석 대상에서 배제해도 분석 결과에 큰 영향을 주지 않음
>* 상관관계가 인과관계를 직접 의미하는 것은 아님
>* 어떤 지역의 병원 수와 자동차 수 사이에 상관관계를 발견해도 병원 수와 자동차<br> 수 사이의 어떤 인과가 존재한다기 보다는 두 속성 다 인구에 대한 인과관계로 보는 것이 타당
>* 상관분석상 강한 상관관계를 나타낸다고 하더라도 중복 속성 여부를 판단할 때는<br> 해당  분야 도메인 지식을 충분히 고려해서 최종 판단하는 것이 바람직함

## 카이제곱 검정(Chi-Square Tests)

* 범주형(이산형) 데이터인 경우, 속성 A와 B 사이의 상관관계는 <br>피어슨(Pearson)의 카이제곱 $\chi^2$검정에 의해 측정 가능<br>
<br>
* 속성 A가 c개의 범주 값 $ a_1, a_2,\ldots,a_c$를 가진다고 가정
* 속성 B가 r개의 범주 값 $ b_1, b_2,\ldots, b_r$를 가진다고 가정<br>


* 속성 A와 B에 의해 구성되는 튜플은 c개의 열과 r개의 행으로 구성되는 분할표로 표현
* <p>($A_i$, $B_j$)를 속성 A가 $a_i$를 가지고, 속성 B가 $b_j$를 가지는 튜플이라고 할 때 $\chi^2$ 의 정의
<div> </div></p>

* $o_{ij}$ : ($A_i$, $B_j$)에 대한 관측도수(observed frequency); 실제로 존재하는 ($A_i$, $B_j$) 튜플 수
* $e_{ij}$ : ($A_i$, $B_j$)에 대한 기대도수(expected frequency); 확률적으로 기대되는 ($A_i$, $B_j$) 튜플 수

<div> </div>

* 기대도수 ejj 계산식
<div> </div>


> * N: 데이터 튜플수
> * count(A = $a_i$) : 속성 A에 대하여 $a_i$를 갖는 튜플 수
> * count(B = $b_j$}) : 속성 B에 대하여 $b_j$를 갖는 튜플 수

* $\chi^2$통계량은 속성 A와 B가 독립이라는 가설을 검증
* 카이제곱 검정은 자유도 (r- 1) x (c -1)을 갖는 유의수준에 근거하여 검정
