# Feature Engineering 정의

* 머신러닝 알고리즘을 작동하기 위해 데이터에 대한 도메인 <br> 지식을 활용하여 특징(Feature)를 만들어내는 과정 <br>
 <br>
* 머신러닝 모델을 위한 데이터 테이블의 컬럼(특징)을 생성하거나  <br>선택하는 작업을 의미 <br>
 <br>
* 모델의 성능을 높이기 위해 모델에 입력할 데이터를 만들기 위해 <br> 주어진 초기 데이터로부터 특징을 가공하고 생성하는 전체 과정을 의미 <br>
 <br>
* Feature Engineering은 모델 성능에 미치는 영향이 크기 때문에 <br> 머신러닝 응용에 있어서굉장히 중요한 단계이며,  <br>전문성과 시간과 비용이 많이 드는 작업

# 특징 선택(Feature Selection)

* 특징 랭킹(Feature Ranking) 또는 특징 중요도(Feature Importance)라고도 불림 <br>
 <br>
* 분류 모델 중 Decision Tree 같은 경우는 트리의 상단에 있을 수록  <br>중요도가 높으므로 이를 반영하여 특징 별로 중요도를 매길수 있음 <br>
 <br>
* 회귀 모델의 경우 forward selection과 backward elimination 같은 알고리즘을 통해 특징을 선택 <br>
 <br>
 

# 차원 감소(Dimension Reduction)

* 차원 감소는 feature extraction 다시 말해 특징 추출이라는 말로도 불림 <br>
 <br>
* 차원축소는 단순히 데이터의 압축이나 잡음(noise)을 제거하는 것이 아님 <br>
 <br>
* 물론 차원축소로 데이터의 압축이나 잡음을 제거하는 효과도 있겠지만, <br> 이것의 가장 중요한 의의는 관측 데이터를 잘 설명할 수 있는 잠재 공간(latent space)을 찾는 것 <br>
 <br>
* 가장 대표적인 알고리즘에 PCA(Principle Component Analysis)가 있음 <br>
 <br>
* PCA를 간단히 설명하면 각 변수(Feature)를 하나의 축으로 투영시켰을 때  <br>분산이 가장큰 축을 첫번째 주성분으로 선택하고 그 다음 큰 축을 두번째 주성분으로 <br> 선택하고 데이터를 선형 변환하여 다차원을 축소하는 방법

# 특징 생성(Feature Generation)

* 특징 구축(Feature Construction)이라고도 하며, 이 방법을  <br>흔히 많은 사람들이 FeatureEngineering이라고 말함 <br>
 <br>
* 초기에 주어진 데이터로부터 모델링 성능을 높이는 새로운 특성을 만드는 과정 <br>

* 데이터에 대한 도메인(분야) 전문성을 바탕으로 데이터를 합치거나 <br> 쪼개는 등의 작업을거쳐 새로운 Feature를 만들게 됨 <br>
 <br>
* 간단한 예로 시간 데이터를 AM / PM 으로 나누는 것이 있음 <br>
 <br>
* 이 작업은 한번 해서 끝나는 것이 아니라 끊임 없이 모델링 성능을 높이는  <br>목적으로 반복해서 작업할 수 있는 부분이기 때문에 전문성과 경험에 따라 <br> 비용과 시간을 줄일 수 있는 부분

## 관점에 따른 Features 분류

1. 비즈니스 기반 특징(Business driven features):
    * 해결하려는 문제가 있는 현장인 비즈니스관점에서 데이터를 분석하여 특징을 만들어내는 관점

2. 데이터 기반 특징(Data driven features) : 
    * 비즈니스 관점이 없어도 주어진 데이터를 다루는 과정에서 특징을 만들어내는 관점   
<br>
* **실제로는 두 가지 방법이 분리되어 있지 않고 서로 조합되어 좋은 Feature를 만들게 됨**

## Feature Engineering 방법 분류

<div>  </div>

### 에러 분석(Error Analysis - Post-Modeling)

* Start with larger errors : <br>
    * 모델을 통해 나온 모든 값을 확인하기 보다 '에러(Error)'값이 큰 feature부터 확인하는 방법<br>
<br>
* Segment By classes :
    * 평균 에러(Error)값을 기준으로 Segment를 나누어 비교하면서 분석하는 방법<br>
    <br>

* Unsupervised clustering : 
    * 패턴을 발견하는데 어려움이 있을 경우 Unsupervised(비지도)학습인 <br>clustering 알고리즘을 사용하여 분류되지 않은 값들을 확인하는 방법. <br>여기서clustering을 클래스로 분류하는 것이 아니라 <br>에러(Error)의 원인을 찾는 방법으로 사용해야 한다는 점을 주의<br>
<br>
* Ask colleagues or domain experts :
    * 데이터를 통해서 발견할 수 없다면 도메인(분야) 전문가의 도움을 통해<br> 에러(Error)의 원인을 찾아낼 수도 있음
